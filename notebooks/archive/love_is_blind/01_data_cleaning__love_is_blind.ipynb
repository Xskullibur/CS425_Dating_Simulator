```python
def parse_file(file_path):
    file = open(file_path, "r", encoding="utf-8")
    html = file.read()

    soup = BeautifulSoup(html, "html.parser")

    for script_tag in soup.find_all("script"):
        script_tag.decompose()

    for ins_tag in soup.find_all("ins"):
        ins_tag.decompose()

    script = list(soup.find("div", class_="full-script").children)

    # Clean the script
    script = [element for element in script if str(element) != "<br/>"]
    script = [element for element in script if "\n" not in str(element)]
    script = [
        str(element).removeprefix("- ") if str(element).startswith("- ") else str(
            element) for element in script]
    script = [element for element in script if "_" not in str(element)]

    cleaned_script = []
    for element in script:
        element_str = str(element)
        if element_str and element_str[0].islower() and cleaned_script:
            cleaned_script[-1] += " " + element_str
        else:
            cleaned_script.append(element_str)

    final_script = []
    i = 0
    while i < len(cleaned_script):
        element_str = str(cleaned_script[i])
        if element_str.startswith("..."):
            found = False
            for j in range(len(final_script) - 1, -1, -1):
                prev_element_str = str(final_script[j])
                if prev_element_str.endswith("..."):
                    merged = prev_element_str.rstrip(
                        "...").rstrip() + " " + element_str.lstrip("...").lstrip()
                    final_script[j] = merged
                    found = True
                    break
            if not found:
                final_script.append(element_str)
            i += 1
        else:
            final_script.append(element_str)
            i += 1

    # Find the first element that contains "♪" and join it with the subsequent element that also contains "♪" repeatedly
    i = 0
    while i < len(final_script) - 1:
        element_str = str(final_script[i])
        if "♪" in element_str:
            j = i + 1
            while j < len(final_script):
                next_element_str = str(final_script[j])
                if "♪" in next_element_str:
                    # Join them
                    merged = element_str + " " + next_element_str
                    final_script[i] = merged
                    del final_script[j]
                else:
                    break
        i += 1

    # Remove all elements that has a "♪" prefix and suffix
    final_script = [element for element in final_script if
          not (str(element).strip().startswith("♪") and str(element).endswith("♪"))]

    # Find all elements that contains "♪" and print them
    for i in range(len(final_script)):
        element_str = str(final_script[i])
        if "♪" in element_str:
            print(f"Element {i} contains '♪': {element_str}")
            print()

    # Find all elements that contains text with the following pattern "[*]" and print them
    import re
    pattern = r"\[.*?\]"
    # Replace text with the following pattern "[*]" with an empty string
    final_script = [re.sub(pattern, "", str(element)).strip() for element in final_script]

    # Remove empty strings
    final_script = [element for element in final_script if element]

    for i in range(len(final_script)):
        element_str = str(final_script[i])
        if re.search(pattern, element_str):
            print(f"Element {i} contains pattern '[*]': {element_str}")
            print()

    return final_script
```
